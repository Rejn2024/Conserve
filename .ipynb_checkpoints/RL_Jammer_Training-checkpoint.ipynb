```python
from __future__ import annotations
from dataclasses import dataclass
from typing import Any, Dict, Optional, Tuple
import numpy as np
import torch
import gymnasium as gym
from gymnasium import spaces

import advanced_link_skdsp_v6_robust as link6
import score_iq_decode as scorer


@dataclass
class JammerEnvConfig:
    jammer_sampling_freq: float = 2e9
    max_steps: int = 1                # start with contextual bandit (1-step episodes)
    power_penalty_weight: float = 0.01
    metric_div_weight: float = 0.1
    decode_weight: float = 1.0
    device: str = "cuda"


class JammerEnv(gym.Env):
    """
    One episode == one training sample from cached dataset.
    State uses precomputed iq1/iq2/iq3 + summary features from whole IQ/meta.
    Action parameterizes jammer controls; controller converts action -> tx_iq.
    """

    metadata = {"render_modes": []}

    def __init__(
        self,
        sample_provider,   # callable or iterator returning one cached sample dict
        jammer_controller, # callable(sample, action)-> tx_iq at jammer_sampling_freq
        cfg: JammerEnvConfig,
    ):
        super().__init__()
        self.sample_provider = sample_provider
        self.jammer_controller = jammer_controller
        self.cfg = cfg
        self.device = torch.device(cfg.device)

        # Example: continuous action vector (adapt dims to controller API)
        # [tone_freq_norm, tone_amp, pulse_width_norm, duty_cycle, chirp_rate_norm, ...]
        self.action_dim = 8
        self.action_space = spaces.Box(
            low=-1.0, high=1.0, shape=(self.action_dim,), dtype=np.float32
        )

        # Example compact observation vector (replace with richer encoding if needed)
        # Could include summary stats of iq1/iq2/iq3 + metadata-derived values.
        self.obs_dim = 64
        self.observation_space = spaces.Box(
            low=-np.inf, high=np.inf, shape=(self.obs_dim,), dtype=np.float32
        )

        self._sample: Optional[Dict[str, Any]] = None
        self._step_idx = 0

    def _build_obs(self, sample: Dict[str, Any]) -> np.ndarray:
        # Minimal placeholder featurization (replace with your preferred frontend):
        iq1 = sample["iq1"]
        iq2 = sample["iq2"]
        iq3 = sample["iq3"]

        def summarize(x: torch.Tensor):
            x = x.detach().cpu()
            p = torch.mean(torch.abs(x) ** 2).item()
            r = torch.mean(x.real).item()
            i = torch.mean(x.imag).item()
            s = torch.std(torch.abs(x)).item()
            return [p, r, i, s]

        feats = summarize(iq1) + summarize(iq2) + summarize(iq3)

        meta = sample["whole_meta"]
        feats += [
            float(meta.get("sample_rate_hz", 0.0)),
            float(meta.get("carrier_hz", 0.0)),
            float(meta.get("rf_center_hz", 0.0)),
            float(meta.get("sps", 0.0)),
            float(meta.get("beta", 0.0)),
            float(meta.get("span", 0.0)),
        ]

        # pad/truncate to obs_dim
        arr = np.array(feats, dtype=np.float32)
        if arr.size < self.obs_dim:
            arr = np.pad(arr, (0, self.obs_dim - arr.size))
        else:
            arr = arr[: self.obs_dim]
        return arr

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self._step_idx = 0
        self._sample = self.sample_provider()  # expects cached sample dict
        obs = self._build_obs(self._sample)
        info = {"sample_name": self._sample.get("sample_name", None)}
        return obs, info

    def step(self, action):
        assert self._sample is not None, "Call reset() first"
        self._step_idx += 1

        sample = self._sample
        whole_iq = sample["whole_iq"].to(self.device)
        whole_meta = sample["whole_meta"]
        whole_sr = float(sample["whole_sample_rate_hz"])

        # 1) action -> jammer tx IQ at jammer sampling rate
        tx_iq = self.jammer_controller(sample, action, device=self.device)

        # 2) resample jammer to whole sample rate and compose jammed signal
        jam_resam = link6.resample_iq(tx_iq, self.cfg.jammer_sampling_freq, whole_sr)
        if jam_resam.shape[0] < whole_iq.shape[0]:
            reps = int(np.ceil(whole_iq.shape[0] / jam_resam.shape[0]))
            jam_resam = jam_resam.repeat(reps)
        jam_resam = jam_resam[: whole_iq.shape[0]]
        jammed = whole_iq + jam_resam.to(dtype=torch.complex64)

        # 3) decode and score
        rx_result = link6.rx_command_iq(jammed, whole_meta)
        decode_score = float(scorer.score_decode(rx_result, whole_meta))
        metric_div = float(rx_result.get("metric_div", 0.0)) if rx_result is not None else 0.0

        # 4) penalties
        jam_power = torch.mean(torch.abs(jam_resam) ** 2).item()

        reward = (
            - self.cfg.decode_weight * decode_score
            - self.cfg.metric_div_weight * metric_div
            - self.cfg.power_penalty_weight * jam_power
        )

        terminated = self._step_idx >= self.cfg.max_steps
        truncated = False
        obs = self._build_obs(sample)  # bandit-like; optionally include last action/reward
        info = {
            "decode_score": decode_score,
            "metric_div": metric_div,
            "jam_power": jam_power,
            "rx_result": rx_result,
        }
        return obs, reward, terminated, truncated, info
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Normal
import numpy as np


class ActorCritic(nn.Module):
    def __init__(self, obs_dim: int, act_dim: int, hidden: int = 256):
        super().__init__()
        self.body = nn.Sequential(
            nn.Linear(obs_dim, hidden), nn.Tanh(),
            nn.Linear(hidden, hidden), nn.Tanh(),
        )
        self.mu_head = nn.Linear(hidden, act_dim)
        self.log_std = nn.Parameter(torch.zeros(act_dim))
        self.v_head = nn.Linear(hidden, 1)

    def forward(self, obs):
        h = self.body(obs)
        mu = self.mu_head(h)
        std = torch.exp(self.log_std).expand_as(mu)
        v = self.v_head(h).squeeze(-1)
        return mu, std, v

    def get_action_value_logp(self, obs, action=None):
        mu, std, v = self(obs)
        dist = Normal(mu, std)
        if action is None:
            action = dist.rsample()
        logp = dist.log_prob(action).sum(-1)
        entropy = dist.entropy().sum(-1)
        return action, logp, entropy, v


def compute_gae(rewards, values, dones, gamma=0.99, lam=0.95):
    # rewards, values, dones: [T]
    adv = torch.zeros_like(rewards)
    lastgaelam = 0.0
    for t in reversed(range(len(rewards))):
        next_nonterminal = 1.0 - dones[t]
        next_value = values[t + 1] if t + 1 < len(values) else 0.0
        delta = rewards[t] + gamma * next_value * next_nonterminal - values[t]
        lastgaelam = delta + gamma * lam * next_nonterminal * lastgaelam
        adv[t] = lastgaelam
    returns = adv + values
    return adv, returns


def train_ppo(
    env,
    total_updates=500,
    rollout_steps=256,
    epochs=10,
    minibatch_size=64,
    gamma=0.99,
    lam=0.95,
    clip_coef=0.2,
    vf_coef=0.5,
    ent_coef=0.01,
    lr=3e-4,
    max_grad_norm=0.5,
    device="cuda",
):
    obs_dim = env.observation_space.shape[0]
    act_dim = env.action_space.shape[0]
    model = ActorCritic(obs_dim, act_dim).to(device)
    opt = optim.Adam(model.parameters(), lr=lr)

    obs, _ = env.reset()
    obs = torch.tensor(obs, dtype=torch.float32, device=device)

    for update in range(1, total_updates + 1):
        obs_buf, act_buf, logp_buf = [], [], []
        rew_buf, done_buf, val_buf = [], [], []

        for _ in range(rollout_steps):
            with torch.no_grad():
                action, logp, _, value = model.get_action_value_logp(obs.unsqueeze(0))
            action_np = action.squeeze(0).cpu().numpy().astype(np.float32)

            next_obs, reward, terminated, truncated, info = env.step(action_np)
            done = float(terminated or truncated)

            obs_buf.append(obs)
            act_buf.append(action.squeeze(0))
            logp_buf.append(logp.squeeze(0))
            rew_buf.append(torch.tensor(reward, dtype=torch.float32, device=device))
            done_buf.append(torch.tensor(done, dtype=torch.float32, device=device))
            val_buf.append(value.squeeze(0))

            if done:
                next_obs, _ = env.reset()

            obs = torch.tensor(next_obs, dtype=torch.float32, device=device)

        # Stack rollout tensors
        obs_t = torch.stack(obs_buf)
        act_t = torch.stack(act_buf)
        logp_t = torch.stack(logp_buf)
        rew_t = torch.stack(rew_buf)
        done_t = torch.stack(done_buf)
        val_t = torch.stack(val_buf)

        with torch.no_grad():
            adv_t, ret_t = compute_gae(rew_t, val_t, done_t, gamma=gamma, lam=lam)
            adv_t = (adv_t - adv_t.mean()) / (adv_t.std() + 1e-8)

        # PPO updates
        n = rollout_steps
        idx = np.arange(n)
        for _ in range(epochs):
            np.random.shuffle(idx)
            for start in range(0, n, minibatch_size):
                mb = idx[start:start + minibatch_size]
                mb_obs = obs_t[mb]
                mb_act = act_t[mb]
                mb_old_logp = logp_t[mb]
                mb_adv = adv_t[mb]
                mb_ret = ret_t[mb]

                _, new_logp, entropy, value = model.get_action_value_logp(mb_obs, mb_act)
                ratio = torch.exp(new_logp - mb_old_logp)

                pg_loss1 = -mb_adv * ratio
                pg_loss2 = -mb_adv * torch.clamp(ratio, 1 - clip_coef, 1 + clip_coef)
                pg_loss = torch.max(pg_loss1, pg_loss2).mean()

                v_loss = 0.5 * ((value - mb_ret) ** 2).mean()
                ent_loss = entropy.mean()

                loss = pg_loss + vf_coef * v_loss - ent_coef * ent_loss

                opt.zero_grad()
                loss.backward()
                nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)
                opt.step()

        if update % 10 == 0:
            print(f"update={update} reward_mean={rew_t.mean().item():.4f} "
                  f"decode~={info.get('decode_score', np.nan):.3f}")

    return model
```
